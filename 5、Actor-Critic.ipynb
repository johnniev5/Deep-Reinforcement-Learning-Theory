policy gradient存在估计reward的不稳定性，太大太小，都会影响概率的sample，那么我们希望能够习得一个分布，使得这个满足一个分布，该如何做呢？

仔细想想，policy gradient是基于获得最大的reward为目的的策略优化方法，而q learning则是基于在某个state s，期望采取某个action来使得得到的value值最大，这两种思想本质上是一致的，都是期望最大，最直观的想法就是将获得最大reward换成得到最大value，直观上reward换成q function，baseline b换成value function...





直观上就是加入Q funciton，它能够习得分布...





注意：这里直接将期望换成了无期望的式子，实际期望应该是真实的数据，而这边使用sample的数据作为q function输出...
