on-policy:

要learn的policy与跟环境互动的是同一个（自己玩，从玩中学习）
    
off-policy:    

要learn的policy与跟环境互动的不是同一个（看别人玩，学习别人的方法）



该如何直观地去理解off-policy呢？

比如你现在在学习一样东西，并希望通过学习后能够实际应用到操作中？方法有在实际操作中学习，而另一种方法就是通过学习他人，然后再应用到实际操作中，而off-policy就是这种方法。它是从别人那里学习多种方法，然后应用到实际，感觉不够了，再去从别人那里学习，唯一跟自学自用的方法是，花费的时间比较短，效率高，因为自学的话，都需要自己去不断调整可能存在的问题，而跟别人学它是直接习得别人的经验... （感觉举例不是很恰当，但是确是这个意思...）

使用importance sampling存在的问题，即使均值一样，方差有可能是不一样的，那怎么办，该如何衡量两个分布距离呢，很自然就想到了KL距离。
