大部分情况强化学习是得不到reward，这就会导致spare的结果...

reward shaping

人为设定一些reward，使得目标更明确...

加入curiosity... 我们不仅会计算total reward，而且还会有total curiosity...




实现的方法是，如果目标比较难以预测（预测与实际差距越大），我们就鼓励model去冒险，去探索...

单是某个目标是比较难预测时不够的，那会就直接执行这个难以预测的目标，我们还需要加入feature ext，过滤掉没有意义的目标...过滤的方法是使得与前次action无关的过滤掉...



Curriculum Learning - 规划学习

渐进式学习，从简单到难的学习方法.. 比如先让机器人拿起板子，然后再将板子插入到木棍上...



Hierarchical Reinforcement Learning - 层次化强化学习

上层提取目标，下层负责执行...



这些方法都是为了解决传统强化学习中存在的reward为0的情况，这些情况会很多，会浪费很多时间... 我们希望其更加智能...
